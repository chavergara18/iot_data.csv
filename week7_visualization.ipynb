```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import os

# Number of simulated logistics records
num_records = 100

# Empty list for storing generated records
data = []

# Possible shipment statuses
statuses = ["In Transit", "Delivered", "Delayed", "Processing"]

# Generate random logistics tracking data
for _ in range(num_records):
    record = {
        "timestamp": datetime.now() - timedelta(minutes=np.random.randint(0, 1440)),
        "package_id": f"PKG{np.random.randint(1000, 9999)}",
        "rfid_tag": f"RFID{np.random.randint(10000, 99999)}",
        "latitude": round(np.random.uniform(13.0, 15.0), 6),
        "longitude": round(np.random.uniform(120.0, 122.0), 6),
        "temperature": round(np.random.uniform(2.0, 10.0), 1),
        "status": np.random.choice(statuses)
    }

    data.append(record)

# Convert to DataFrame
df = pd.DataFrame(data)

# Show first 5 rows
print(df.head())

# Save CSV and JSON to OneDrive Desktop
csv_path = r"C:\Users\Cha\OneDrive\Desktop\iot_Data.csv"
json_path = r"C:\Users\Cha\OneDrive\Desktop\iot_Data.json"

df.to_csv(csv_path, index=False)
df.to_json(json_path, orient="records")

# Verify file creation
print("\nFiles saved successfully!")

print("\nCSV Exists:", os.path.exists(csv_path))
print("JSON Exists:", os.path.exists(json_path))

print("\nCSV Location:")
print(csv_path)

print("\nJSON Location:")
print(json_path)

# Dataset information
print("\nDataset Information:")
print(df.info())

# Statistical summary
print("\nStatistical Summary:")
print(df.describe())

# Temperature graph
plt.figure(figsize=(10,5))
plt.plot(df["temperature"])
plt.title("Temperature Sensor Readings During Transport")
plt.xlabel("Record Number")
plt.ylabel("Temperature (°C)")
plt.show()

# Shipment status graph
status_counts = df["status"].value_counts()

plt.figure(figsize=(8,5))
status_counts.plot(kind="bar")

plt.title("Shipment Status Distribution")
plt.xlabel("Status")
plt.ylabel("Number of Packages")

plt.show()

# GPS scatter plot
plt.figure(figsize=(8,6))
plt.scatter(df["longitude"], df["latitude"])

plt.title("GPS Shipment Locations")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

plt.show()
```

```python
import pandas as pd

total_records = contract.functions.getTotalRecords().call()

data = []

for i in range(total_records):
    record = contract.functions.getRecord(i).call()

    data.append({
        "timestamp": record[0],
        "device_id": record[1],
        "data_type": record[2],
        "data_value": record[3]
    })

df = pd.DataFrame(data)

df["timestamp"] = pd.to_datetime(df["timestamp"], unit="s")

print(df)
```

```python
df = df.drop_duplicates(
    subset=["device_id", "data_type", "data_value"],
    keep="first"
)
```

```python
import numpy as np

df["numeric_value"] = (
    df["data_value"]
    .str.extract(r'(\d+\.?\d*)')
    .astype(float)
)

df.fillna(0, inplace=True)

print(df)
```

```python
df.to_csv("cleaned_iot_data.csv", index=False)

print("✅ Cleaned IoT data saved successfully as cleaned_iot_data.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned IoT data
df = pd.read_csv("cleaned_iot_data.csv")

# Display first few rows
print(df.head())

# Convert timestamp column to datetime
df["timestamp"] = pd.to_datetime(df["timestamp"])

# Set visualization style
sns.set(style="whitegrid")

# Create line plot
plt.figure(figsize=(12, 6))

sns.lineplot(
    x=df["timestamp"],
    y=df["numeric_value"],
    hue=df["data_type"],
    marker="o"
)

# Rotate timestamps
plt.xticks(rotation=45)

# Add labels and title
plt.title("IoT Sensor Readings Over Time", fontsize=14)
plt.xlabel("Timestamp", fontsize=12)
plt.ylabel("Sensor Value", fontsize=12)

# Show legend
plt.legend(title="Sensor Type")

# Adjust layout
plt.tight_layout()

plt.legend(title="Sensor Type")

plt.savefig("iot_sensor_readings.png", bbox_inches="tight")

# Display graph
plt.show()
```
